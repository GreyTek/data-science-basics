```python
import matplotlib.pyplot as plt
import numpy as np
x =np.arange(0,100)
y = x*2
z = x**2
```

```python
%matplotlib inline
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Exercise')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.2,0.5,0.2,0.2])

axes1.plot()
axes2.plot()
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.2,0.5,0.4,0.4])

# Bigger Plot
axes1.plot(x,z)
axes1.set_title('Bigger Plot')
axes1.set_xlabel('X')
axes1.set_ylabel('Y')


#Smaller Plot
axes2.plot(x,y)
axes2.set_title('Smaller Plot')
axes2.set_xlabel('X')
axes2.set_ylabel('Y')
axes2.set_xlim([20,22])
axes2.set_ylim([30,50])
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2,figsize=(12,2))
axes[0].plot(x,y,ls='--',color='blue',lw=2)
axes[1].plot(y,z,lw=3,color='red')
```
