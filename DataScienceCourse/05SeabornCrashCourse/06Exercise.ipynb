```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style('whitegrid')
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
titanic['embark_town'].unique()
```

```python
sns.jointplot(x='fare',y='age',data=titanic)
```

```python
sns.barplot(x='age',y='alive',data=titanic,hue='sex')
```

```python
sns.boxplot(x='age', y='alive', data=titanic, hue='sex')
```

```python
sns.displot(titanic['fare'],bins=20,color='orange')
```

```python
sns.boxplot(x='class', y='age', data=titanic,hue='sex')
```

```python
sns.swarmplot(x='class',y='age',data=titanic,hue='sex')
```

```python
sns.countplot(x='sex',data=titanic)
```

```python
tit = titanic.corr()
```

```python
sns.heatmap(tit)
```
